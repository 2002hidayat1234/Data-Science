```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df.head(20)
```

```python
#Cek data yang null
df.isnull().sum()
```

```python
#Atribut tabel untuk informasi data pasien
df.info()
```

```python
#Buat prediksi dengan data uji
from sklearn.model_selection import train_test_split

X = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
Y = df['target']
```

```python
#Memisahkan train dan test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Training set
print(X_train)
print(Y_train)
```

```python
#Test set
print(X_test)
print(Y_test)
```

```python
#menggunakan naive bayes
from sklearn.naive_bayes import GaussianNB

#Latih model data dengan naive baiyes
model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
#Jumlah pasien yang terkena atau tidak terkena penyakit jantung
hitung_target = df['target'].value_counts()
print(hitung_target)
```

```python
#Membuat prediksi dengan ata uji di atas
Y_prediction = model.predict(X_test)


print(Y_prediction)
print(Y_test)
```

```python
#Mengevaluasi keakuratan model
from sklearn.metrics import accuracy_score
akurasi = accuracy_score(Y_test, Y_prediction)
print("Akurasi yang didapatkan: ", akurasi)
```

```python
#Membuat confusion matrics
from sklearn.metrics import confusion_matrix as cm

#Confusion matrics yang didapatkan
cm(Y_prediction, Y_test)  
# TP  FP
# FN  TN
```

```python
#Membuat laporan klasifikasi
from sklearn.metrics import classification_report as cr
overall_report = cr(Y_test, Y_prediction)
print(overall_report)
```
